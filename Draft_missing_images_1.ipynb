I accidentally deleted some images on google drive. This notebook is to find out how many were deleted and to restore them.

```python
import os
from datetime import datetime
import glob
from glob import glob
import pandas as pd
import PIL
from PIL import Image
```

```python
drive_path = '/content/drive/MyDrive/'
```

**To check which files were modified on a specific date:**

```python
target_date_str = '2024-02-04'
target_date = datetime.strptime(target_date_str, '%Y-%m-%d')
```

```python
target_date
```

```python
for foldername, subfolders, filenames in os.walk(drive_path, target_date):
    for filename in filenames:
        file_path = os.path.join(foldername, filename)
        # Get the last modification time of the file
        modified_time = datetime.fromtimestamp(os.path.getmtime(file_path))
        if modified_time.date() == target_date.date():
            print(f'{file_path}')
```

```python
def count_files_in_directory(directory_path):

    file_count = 0
    for foldername, subfolders, filenames in os.walk(directory_path):
        file_count += len(filenames)

    return file_count
```
